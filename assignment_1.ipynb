```python
class Cost:
    def __init__(self, rate, name, amount=1, increment=1):
        """Costs class. Pass parameters and use methods to return company costs and such.

        Args:
            rate (int): The hourly rate or flat rate.
            amount (int): Amount of software or staff.
            name (str): Name of the cost
            increment (int, optional): Null or 1 if flat rate. If employee works 50 hours, enter 50 hours.
            Defaults to 1.
        """
        self.rate = rate
        self.name = name
        self.amount = amount
        self.increment = increment

    def get_cost(self):
        """Returns the cost"""
        cost = (self.rate * self.amount) * self.increment

        return cost

    def print_cost(self):
        """returns the cost in a string"""
        cost_str = f"{self.name}: {self.get_cost()}"
        
        return cost_str
```
